```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('wisc_bc_data.csv')
data.head()
```

## To change the data type of a category !

```python
data.dtypes
data['diagnosis'] =data.diagnosis.astype('category') 
data.dtypes
```

```python
data.head()
```

```python
X = data.drop('diagnosis', axis = 1)
y = data['diagnosis']
X.head()
```

```python
X.drop('id',axis = 1,inplace = True)
X.head()
```

## When we have big number difference between our data instances, we need to scale the data
## otherwise the algorithm will think that some columns are more important than others,when
## in this case they all play an equally important part!!
## WE MUST MAKE SURE THAT BEFORE APPLYING THIS WE SEPARATED THE DEPENDENT VARIABLE FROM THE INDEPENDENT VARIABLES!!!

```python
from scipy.stats import zscore

X_scaled = X.apply(zscore)
X_scaled.describe()
```

## We notice that now the std of all columns is equal!

```python
X = pd.DataFrame(X_scaled)
X.head()
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,y,test_size = 0.3, random_state = 1)
```

## We start building the KNN model

```python
from sklearn.neighbors import KNeighborsClassifier

KNN = KNeighborsClassifier(n_neighbors = 5, weights = 'distance')
```

```python
KNN.fit(x_train,y_train)
```

```python
Predict = KNN.predict(x_test)
Predict
```

```python
KNN.score(x_test,y_test)
```

```python
from sklearn import metrics

matrix = metrics.confusion_matrix(y_test,Predict, labels = ['M','B'])
FinalMatrix = pd.DataFrame(matrix, index = [i for i in ['M','B']], columns =[ i for i in ['Predicted M','Predicted B']])


plt.figure(figsize = (15,15))
sns.heatmap(FinalMatrix, annot = True)
```

# Time to find out how to pick the ideal K to have the best score

```python
from sklearn.model_selection import cross_val_score

MyModelPredictions = []

for i in range(1,50):
    
    
    Model =  KNeighborsClassifier(n_neighbors = i)
    score = cross_val_score(Model,X,y, cv =10)
    MyModelPredictions.append(score.mean())

```

```python
MyModelPredictions
```

```python
plt.figure(figsize = (15,5))
plt.plot(range(1,50), MyModelPredictions,color='blue', linestyle='dashed', marker='o',markerfacecolor='red', markersize=10)
plt.title('Accuracy Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Accuracy Rate') 
```

# As we can see from this representation,the highest accuracy score is obtained when k = 9,in which 
# case the accuracy score is around 97%.Let's test this.

```python
MyIdealModel = KNeighborsClassifier(n_neighbors = 9)
MyIdealModel.fit(x_train,y_train)
MyIdealModel.score(x_test,y_test)
```
